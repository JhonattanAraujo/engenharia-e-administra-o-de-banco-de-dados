# criando o df com select


SELECT
    COD_DP,
    NM_DP,
    Endereço,
    Responsavel
FROM
(
    SELECT 
        a.COD_DP,
        a.NM_DP,
        a.Endereço,
        b.Responsavel,
        MAX(a.Data_Carga) AS MaxData_Carga
    FROM DP_Dados a
    JOIN DP_Resp b 
    ON a.COD_DP = b.COD_DP
    GROUP BY
        a.COD_DP,
        a.NM_DP,
        a.Endereço,
        b.Responsavel
) a;


```python
qry = '''
    SELECT
        COD_DP,
        NM_DP,
        Endereço,
        Responsavel
    FROM
    (
        SELECT 
            a.COD_DP,
            a.NM_DP,
            a.Endereço,
            b.Responsavel,
            MAX(a.Data_Carga) AS MaxData_Carga
        FROM DP_Dados a
        JOIN DP_Resp b 
        ON a.COD_DP = b.COD_DP
        GROUP BY
            a.COD_DP,
            a.NM_DP,
            a.Endereço,
            b.Responsavel
    ) a;

'''
dwDP = pd.read_sql(qry,conexaoODS)

conexaoODS.commit()

conexaoODS.close()

dwDP
```
